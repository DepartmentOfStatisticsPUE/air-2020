```
import pandas as pd
import numpy as np
import seaborn as sns
```

```
gosp = pd.read_excel(io = "gospodarstwa.xlsx")
gosp.head()
gosp2 = gosp.dropna()
gosp2.head()
```

Wizualizujemy dane z wykorzystaniem funkcji distplot


```
sns.distplot(a = gosp2["wydg"], 
             kde = False,  ## kernel density estimation
             rug = True,
             hist_kws = dict(edgecolor = "k"))
```

```
wyk = sns.distplot(a = gosp2["wydg"], 
             kde = False,  ## kernel density estimation
             rug = True,
             hist_kws = dict(edgecolor = "k"))
wyk.set_xscale("log")
```

Tworzymy wykres pudełkowy (boxplot)

```
sns.boxplot(x = "d348", y = "wydg", data = gosp2, hue = "klm")
```

Tworzymy wykres pudełkowy ale dodajemy transformację osi Y. Użyjemy logartymu

```
wykres = sns.boxplot(x = "klm", y = "wydg", data = gosp2)
wykres.set_yscale("log")
```

Dodajemy do wykresu pudełkowego wykres punktowy (stripplot)

```
wyk1 = sns.stripplot(x = "klm", y = "wydg", data = gosp2, size = 1)
wyk1 = sns.boxplot(x = "klm", y = "wydg", data = gosp2)
wyk1.set_yscale("log")
```

Badanie zależności między dwiema zmiennymi ciągłymi -- scatterplot, pairplot

```
sns.scatterplot(x = "dochg", y = "wydg", data = gosp2)
sns.regplot(x = "dochg", y = "wydg", data = gosp2)
```

```
sns.pairplot(data = gosp2, vars = ["dochg", "wydg", "d36"])
sns.pairplot(data = gosp2, vars = ["dochg", "wydg", "d36"],  kind = "reg") ## dodajemy regresje
```

```
m = sns.scatterplot(x = "dochg", y = "wydg", data = gosp2)
m.set_yscale("log")
m.set_xscale("log")
```

```
gosp2["wydg_log"]=np.log(gosp2.wydg)
gosp2["dochg_log"]=np.log(gosp2.dochg)
sns.scatterplot(x = "dochg_log", y = "wydg_log", data = gosp2)
```
