```
import pandas as pd
import numpy as np
```

```
## wczytujemy dane
gosp = pd.read_excel(io = "gospodarstwa.xlsx").dropna()
gosp2 = gosp.loc[gosp.dochg > 0]
gosp2.head()
```

Tworzymy tabelę przestawną z funkcją pd.pivot_table

```
pd.pivot_table(data = gosp2, values = "dochg", index = "klm", aggfunc = 'mean')
```

```
pd.pivot_table(data = gosp2, 
               values = ["dochg", "wydg"], 
               index = ["klm", "d21"], 
               aggfunc = ['mean', "median"])
```

Przetwarzamy dane do postaci ramki danych (dataframe)

```
tab1 = pd.pivot_table(data = gosp2, 
               values = ["dochg", "wydg"], 
               index = ["klm", "d21"], 
               aggfunc = ['mean', "median"])

tab1.stack().reset_index()
```
